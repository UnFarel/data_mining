3. Используя библиотеку plotly express, создайте интерактивный график линий с помощью функции [`plotly.express.line`](https://plotly.com/python-api-reference/generated/plotly.express.line.html).
4. В качестве параметров функции `line` передайте массив месяцев в качестве оси `x`, массивы температур в качестве оси `y`, а также названия городов в качестве параметра `color`, чтобы разделить линии по группам .
5. Настройте внешний вид графика, добавляя подписи к осям, заголовок графика и легенду [`.update_layout()`](https://plotly.com/python/creating-and-updating-figures/#updating-figure-layouts).

```python
import numpy as np
import plotly.express as px
```

```python
months = np.arange(1, 13)

months
```

```python
moscow = np.round(-10 + 20 * np.sin(np.pi * (months - 3) / 6))
paris = np.round(5 + 15 * np.sin(np.pi * (months - 3) / 6))
new_york = np.round(0 + 25 * np.sin(np.pi * (months - 3) / 6))

print("Темпа в Москве", moscow)

print("Темпа в Париже", paris)

print("Темпа в Нью-Йорке", new_york)
```

```python
data = {
    "Месяц": np.tile(months, 3),
    "Температура, °C": np.concatenate([moscow, paris, new_york]),
    "Город": ["Москва"] * 12 + ["Париж"] * 12 + ["Нью-Йорк"] * 12
}
```

```python
fig = px.line(
    data,
    x="Месяц",
    y="Температура, °C",
    color="Город",
    title='Средняя температура воздуха по месяцам в трех городах',
    color_discrete_map={
        "Москва": "red",
        "Париж": "teal",
        "Нью-Йорк": "purple"
    }
)
```

```python
fig.update_layout(
    xaxis_title="Месяц",
    yaxis_title="Температура, °C",
    legend_title="Город",
    hovermode="x unified",
    xaxis=dict(
        tickmode='array',
        tickvals=months[1::2],
    )
)
```

### Задание 2.6: Создание 3D графиков

2. На основе массивов `X` и `Y` сгенерируйте массив `Z`, используя любое математическое выражение (например, `Z = numpy.sin(X) + numpy.cos(Y)`).
3. Используя Matplotlib, создайте фигуру, а затем объект для трехмерного графика с помощью [`fig.add_subplot(111, projection='3d')`](https://matplotlib.org/3.2.1/tutorials/toolkits/mplot3d.html)`.
4. Отобразите сгенерированные данные в виде поверхности, используя метод `plot_surface`.
5. Настройте визуальные аспекты графика - прозрачность `set_alpha()` и угол обзора `view_init()`.
6. Добавьте подписи к осям и заголовок графика.



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, 
                      cmap='Blues',      
                      edgecolor='none',    
                      alpha=0.8,           
                      rstride=2,           
                      cstride=2)           

ax.view_init(elev=30, azim=60)  

ax.set_title('Трёхмерный график функции Z = sin(X) * cos(Y)', fontsize=14)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

ax.grid(True)

plt.tight_layout()
plt.show()
```

2. Используя библиотеку plotly express, создайте трехмерный график рассеяния с помощью функции `px.scatter_3d`.
3. В качестве параметров функции `scatter_3d` передайте массивы `X`, `Y` и `Z`.
4. Настройте внешний вид графика, изменяя цвет и размер точек `fig.update_traces()`, а также добавляя подписи к осям и заголовок графика `fig.update_layout()`.






```python
import numpy as np
import plotly.express as px

x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
z = np.random.uniform(0, 10, 100)

fig = px.scatter_3d(
    x=x, y=y, z=z,
    title='Трехмерный график рассеяния случайных точек'
)

fig.update_traces(
    marker=dict(
        size=8,  
        color=z,               
        colorscale='turbo',     
        opacity=0.8,            
        line=dict(width=0.5, color='DarkSlateGrey') 
    )
)

fig.update_layout(
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y', 
        zaxis_title='Z',
        xaxis=dict(range=[0, 10]), 
        yaxis=dict(range=[0, 10]),
        zaxis=dict(range=[0, 10])
    ),
    margin=dict(l=0, r=0, b=0, t=30),  
    title_font=dict(size=20)           
)
```

#### Описание модели SIR:

Модель SIR (Susceptible, Infected, Recovered) представляет собой классическую эпидемиологическую модель, которая разделяет популяцию на три основные группы:
- **S (Susceptible)**: Часть популяции, восприимчивая к инфекции, но еще не зараженная.
- **I (Infected)**: Часть популяции, зараженная инфекцией и способная передавать ее другим.
- **R (Recovered)**: Часть популяции, выздоровевшая после инфекции и получившая иммунитет.

#### Математический аппарат модели SIR:

1. **Дифференциальные уравнения модели SIR:**

   - $\frac{dS}{dt} = -\beta \frac{SI}{N}$ — уменьшение числа восприимчивых из-за заражения.
   - $\frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I$ — изменение числа инфицированных, учитывая новые случаи заражения и выздоровление.
   - $\frac{dR}{dt} = \gamma I$ — увеличение числа выздоровевших благодаря выздоровлению инфицированных.

   Где:
   - $N$ — общая численность популяции (постоянная величина).
   - $\beta$ — коэффициент заразности, определяющий вероятность передачи инфекции от инфицированных к восприимчивым.
   - $\gamma$ — коэффициент выздоровления, обратный среднему периоду инфекционности.

```python
import numpy as np
import plotly.express as px
import pandas as pd

class SIRModel:
    def __init__(self, N, beta, gamma, S0, I0, R0, t_max, dt):
        self.N = N
        self.beta = beta
        self.gamma = gamma
        self.S = S0
        self.I = I0
        self.R = R0
        self.t_max = t_max
        self.dt = dt
        self.t = 0
        self.S_list = [S0]
        self.I_list = [I0]
        self.R_list = [R0]
        self.t_list = [0]

    def run(self):
        while self.t < self.t_max:
            dS = -self.beta * self.S * self.I / self.N * self.dt
            dI = (self.beta * self.S * self.I / self.N - self.gamma * self.I) * self.dt
            dR = self.gamma * self.I * self.dt
            self.S += dS
            self.I += dI
            self.R += dR
            self.t += self.dt
            self.S_list.append(self.S)
            self.I_list.append(self.I)
            self.R_list.append(self.R)
            self.t_list.append(self.t)

    def plot(self):
        data = {
            'Время': self.t_list,
            'Восприимчивые': self.S_list,
            'Инфицированные': self.I_list,
            'Выздоровевшие': self.R_list
        }
        df = pd.DataFrame(data)
        fig = px.line(
            df,
            x='Время',
            y=['Восприимчивые', 'Инфицированные', 'Выздоровевшие'],
            title='Динамика модели SIR',
            labels={'value': 'Численность', 'variable': 'Группа'},
            template='plotly_white'
        )
        fig.update_layout(
            xaxis_title='Время (дни)',
            yaxis_title='Количество особей',
            legend_title='Группы населения',
            showlegend=True
        )
        fig.show()
```

```python
import numpy as np
import plotly.express as px
import pandas as pd

class Prey:
    def __init__(self, x0, alpha):
        self.x = x0
        self.alpha = alpha

    def grow(self):
        return self.alpha * self.x

    def die(self, y, beta):
        return -beta * self.x * y

    def update(self, dx):
        self.x += dx

class Predator:
    def __init__(self, y0, gamma):
        self.y = y0
        self.gamma = gamma

    def grow(self, x, delta):
        return delta * x * self.y

    def die(self):
        return -self.gamma * self.y

    def update(self, dy):
        self.y += dy

class EcoSystem:
    def __init__(self, x0, y0, alpha, beta, gamma, delta, t_max, dt):
        self.prey = Prey(x0, alpha)
        self.predator = Predator(y0, gamma)
        self.beta = beta
        self.delta = delta
        self.t_max = t_max
        self.dt = dt
        self.t = 0
        self.x_list = [x0]
        self.y_list = [y0]
        self.t_list = [0]

    def run(self):
        while self.t < self.t_max:
            dx = (self.prey.grow() + self.prey.die(self.predator.y, self.beta)) * self.dt
            dy = (self.predator.grow(self.prey.x, self.delta) + self.predator.die()) * self.dt
            self.prey.update(dx)
            self.predator.update(dy)
            self.t += self.dt
            self.x_list.append(self.prey.x)
            self.y_list.append(self.predator.y)
            self.t_list.append(self.t)

    def plot(self):
        data = {
            'Время': self.t_list,
            'Жертвы': self.x_list,
            'Хищники': self.y_list
        }
        df = pd.DataFrame(data)
        fig_line = px.line(
            df,
            x='Время',
            y=['Жертвы', 'Хищники'],
            title='Динамика популяций хищник-жертва',
            labels={'value': 'Численность', 'variable': 'Группа'},
            template='plotly_white'
        )
        fig_line.update_layout(
            xaxis_title='Время (единицы)',
            yaxis_title='Количество особей',
            legend_title='Группы населения',
            showlegend=True
        )
        fig_phase = px.scatter(
            df,
            x='Жертвы',
            y='Хищники',
            title='Фазовый портрет',
            labels={'Жертвы': 'Численность жертв', 'Хищники': 'Численность хищников'},
            template='plotly_white'
        )
        fig_phase.update_layout(
            xaxis_title='Численность жертв',
            yaxis_title='Численность хищников',
            showlegend=False
        )
        fig_line.show()
        fig_phase.show()
```

```python
# Создаем объект экосистемы с заданными параметрами
ecosystem = EcoSystem(x0=10, y0=5, alpha=0.1, beta=0.01, gamma=0.1, delta=0.01, t_max=1000, dt=0.1)
# Запускаем моделирование
ecosystem.run()
# Визуализируем результаты
ecosystem.plot()
```
